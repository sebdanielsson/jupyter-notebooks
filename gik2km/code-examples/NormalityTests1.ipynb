```python
# Assessing Normality, incl. Shapiro-Wilks Normality test
from numpy.random import seed
from numpy.random import randn
import numpy as np
import scipy.stats
# "Lock" the pseudo-random number generator, to allow replicable analysis
seed(324)
# Return a sample (or samples) from the "standard normal" distribution.
mean = 150
sd = 10 
myData = sd * randn(500) + mean

sampleMean = np.mean(myData)
sampleMedian = np.median(myData)
sampleSD = np.std(myData)

print('mean =',sampleMean)
print('median =',sampleMedian)
print('std deviation =', sampleSD)
#print(myData)
scipy.stats.describe(myData, ddof=1, bias=False)
```

```python
from matplotlib import pyplot
#1. Qualitative assessment: Visual inspection of histogram
# histogram
pyplot.hist(myData)
pyplot.show()
```

```python
#2 QQ plot
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot

# q-q plot
qqplot(myData, line='s')
pyplot.show()
```

```python
from scipy.stats import shapiro
#3. Quantitative assessment: 
#    Shapiro-Wilk test [ scipy.stats.shapiro(x) ]
# H0: The data is drawn from a normal distribution.
# H1: The data is NOT drawn from a normal distribution.
# Significance level
ALPHA = 0.05

# normality test
stat, p = shapiro(myData)
print('Statistics=%.3f, p=%.3f' % (stat, p))
```
