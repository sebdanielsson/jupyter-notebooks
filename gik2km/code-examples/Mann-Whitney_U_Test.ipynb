```python
# Mann-Whitney U test
# H0: There IS NO difference between the distributions of the data samples, i.e. the sample distributions are equal
# Ha: There IS A DIFFERENCE between the distributions of the data samples, i.e. the sample distributions NOT are equal

import matplotlib
from scipy.stats import mannwhitneyu

# two independent samples
# Two separate groups were to assess the quality of products
#   The control group did not have any training on the products
#   The experiment group conducted a one week crash course on how to assess quality of this kind of products 
ctrlGrp = ([11,15, 9, 4,34,17,18,14,12,13,26,31])
expGrp =  ([34,31,35,29,28,12,18,30,14,22,10,29])

#
from matplotlib import pyplot
# Qualitative assessment of normality: 
# Visual inspection of histogram
pyplot.hist(ctrlGrp)
pyplot.show()
pyplot.hist(expGrp)
pyplot.show()

# Visual inspection of boxplots
data = [ ctrlGrp, expGrp ]
fig1, ax2 = matplotlib.pyplot.subplots()
ax2.set_title('Control Group (1) vs Experimental Group (2)')
ax2.boxplot(data)
matplotlib.pyplot.show()


from scipy.stats import shapiro
# Quantitative assessment of normality: 
#    Shapiro-Wilks test [ scipy.stats.shapiro(x) ]
# H0: The data is drawn from a normal distribution.
# H1: The data is NOT drawn from a normal distribution.
# Significance level
ALPHA = 0.05

# normality test
stat, p = shapiro(ctrlGrp)
print(' Shapiro-Wilks test: Statistics=%.3f, p=%.3f' % (stat, p))

# normality test
stat, p = shapiro(expGrp)
print('Shapiro-Wilks test: Statistics=%.3f, p=%.3f' % (stat, p))


# compare samples
stat, p = mannwhitneyu(ctrlGrp, expGrp)
print('Mann-Whitney U: Statistics=%.3f, p=%.3f' % (stat, p))


```
